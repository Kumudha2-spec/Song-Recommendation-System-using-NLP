```python
def make_polite(text):
    text = text.replace("now", "when you have time")
    text = text.replace("ASAP", "as soon as possible")
    text = text.replace("Do it", "Could you please do it")
    text = text.replace("Send", "Please send")
    text = text.replace("!", ".")
    if not text.lower().startswith(("please", "could", "would")):
        text = "Please " + text[0].lower() + text[1:]
    return text
```

```python
# ðŸŽµ Emotion-Based Music Recommender using NLTK

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Step 1ï¸âƒ£ â€“ Preprocessing
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]
    return filtered_tokens

# Step 2ï¸âƒ£ â€“ Emotion Lexicon (Custom dictionary)
emotion_lexicon = {
    'happy': 'joy',
    'joyful': 'joy',
    'glad': 'joy',
    'excited': 'joy',
    'love': 'love',
    'romantic': 'love',
    'heart': 'love',
    'sad': 'sadness',
    'cry': 'sadness',
    'alone': 'sadness',
    'tired': 'sadness',
    'angry': 'anger',
    'furious': 'anger',
    'hate': 'anger',
    'mad': 'anger',
    'scared': 'fear',
    'afraid': 'fear',
    'nervous': 'fear',
    'fear': 'fear',
    'surprised': 'surprise',
    'shocked': 'surprise',
    'amazed': 'surprise'
}

# Step 3ï¸âƒ£ â€“ Emotion detection
def detect_emotion(text):
    tokens = preprocess_text(text)
    emotion_count = {}
    for word in tokens:
        if word in emotion_lexicon:
            emotion = emotion_lexicon[word]
            emotion_count[emotion] = emotion_count.get(emotion, 0) + 1

    if not emotion_count:
        return "neutral"
    # Return emotion with max count
    return max(emotion_count, key=emotion_count.get)

# Step 4ï¸âƒ£ â€“ Music recommendation based on emotion
def recommend_music(emotion):
    music_db = {
        'joy': ["Happy â€“ Pharrell Williams", "Best Day of My Life â€“ American Authors"],
        'love': ["Perfect â€“ Ed Sheeran", "All of Me â€“ John Legend"],
        'sadness': ["Someone Like You â€“ Adele", "Fix You â€“ Coldplay"],
        'anger': ["Believer â€“ Imagine Dragons", "Numb â€“ Linkin Park"],
        'fear': ["Fight Song â€“ Rachel Platten", "Brave â€“ Sara Bareilles"],
        'surprise': ["Good Life â€“ OneRepublic", "Pompeii â€“ Bastille"],
        'neutral': ["Let It Be â€“ The Beatles", "Dreams â€“ Fleetwood Mac"]
    }
    return music_db.get(emotion, ["No songs found"])

# Step 5ï¸âƒ£ â€“ Demo
user_input = input("How are you feeling today? ðŸ§ \n> ")

emotion = detect_emotion(user_input)
songs = recommend_music(emotion)

print("\nðŸŽ¯ Detected Emotion:", emotion.upper())
print("ðŸŽµ Recommended Songs:")
for song in songs:
    print(" -", song)
```
